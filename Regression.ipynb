#Import Libraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("manufacturing_dataset_.xlsx")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

#Display Headings

```python
df.head()
```

# Display information

```python
df.info()
```

# Check for misssing values

```python
df.isnull().sum()
```

#Handle the null by dropping

```python
df = df.dropna(subset=["Production_Cost"])
```

#Define X and Y

```python
y = df["Production_Cost"]

X = df.drop(columns=["Production_Cost","Product_ID"])
```

#Filling missing numerical values

```python
num_cols = X.select_dtypes(include=["int64","float64"]).columns

X[num_cols] = X[num_cols].fillna(X[num_cols].mean())
```

#Filling missing categorical values

```python
cat_cols = X.select_dtypes(include=["object"]).columns

for col in cat_cols:
    X[col] = X[col].fillna(X[col].mode()[0])
```

#Comfirming with a null value is still present

```python
df.isnull().sum()
```

```python
X = pd.get_dummies(X, drop_first=True)
```

#Split into training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
test_size=0.20,random_state=42)
```

#Build the model

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
print(model.intercept_)
```

```python
coefficients = pd.DataFrame({
    "Variable": X.columns,
    "Coefficient": model.coef_
})

coefficients
```

```python
for feature, coef in zip(X.columns, model.coef_):
    print(f"{feature}: {coef}")
```

#Make predictions

```python
y_pred = model.predict(X_test)
```

Evaluate the model

```python
MAE= mean_absolute_error(y_test, y_pred)
```

```python
print("mae:", MAE)
```

```python
MSE = mean_squared_error(y_test, y_pred)
print("mse:", MSE)
```

```python
RMSE= np.sqrt(MSE)
print("rmse:", RMSE)
```

```python
R2 = r2_score(y_test, y_pred)
print("r²:", R2)
```
